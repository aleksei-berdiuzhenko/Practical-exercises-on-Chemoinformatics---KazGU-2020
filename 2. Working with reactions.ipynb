##### All ***CGRTools*** parsers support python contexts

```python
with RDFRead(r'scifinder_export.rdf') as f: #for reading files of reactions
    r = next(f) #select one by one
    rs = f.read() #for getting all in one list
```

```python
rs #all reaction from the file
```

```python
rxn1 = rs[0]
rxn1
```

```python
rxn2 = rs[1]
rxn2
```
