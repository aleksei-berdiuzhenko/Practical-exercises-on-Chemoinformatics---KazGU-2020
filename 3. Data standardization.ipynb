```python
data = []
for sdf_file in ['logBB.sdf']:
    with SDFRead(sdf_file) as f:
        for x in f:
            x.standardize() #standardization of atom groups
            x.kekule() #standardization of Kekule structures for correcting of incorrectly aromatized rings
            x.implicify_hydrogens() #deletion of explicitly shown hydrogens
            x.thiele() #presenting rings into their aromatic form
            data.append(x)
```

```python
#code for checking for duplicates and separating them into different collectors if they are presented
values = defaultdict(list)
for x in data:
    values[x].append(float(x.meta['logBB']))
duplicates = []
uniques = []
for x, y in values.items():
    if len(y) > 1:
        print(x, y)
        duplicates.append(x)
    else:
        uniques.append((x, y[0]))
```
